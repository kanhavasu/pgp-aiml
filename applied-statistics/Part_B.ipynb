1. Read the data set, clean the data and prepare final dataset to be used for analysis.

```python
# Lets import the necessary libraries
import pandas as pd
```

```python
# Let's read the datasets
df=pd.read_csv("Basketball.csv")
```

```python
# Let's check the dtype of the dataset
df.info()
```

```python
# Let's display first 5 rows of the data frame
df.head()
```

```python
# Let's display last 5 rows of the data frame
df.tail()
```

```python
# Let's replace '-' by zero(0) and NaN by zero(0)
df.replace(to_replace='-',value=0,inplace=True)
df.fillna(0)
```

```python
# Let's drop duplicate rows, if any
df.drop_duplicates()
```

```python
# Let's drop one column TeamLaunch
df.drop("TeamLaunch",axis=1,inplace=True)
```

```python
# Let's convert the dtype of the columns with index greater than 1
column=df.columns
df[column[2:11]]=df[column[2:11]].astype('int64')
```

```python
# Let's add a column PercentageOfWinning the games
percentage_of_winning=df["WonGames"]/df["PlayedGames"]
df.insert(loc=len(df.columns),column="PercentageOfWinning",value=percentage_of_winning,allow_duplicates=True)
```

```python
# Let's add a column PercentageOfLosing the games
percentage_of_losing=df["LostGames"]/df["PlayedGames"]
df.insert(loc=len(df.columns),column="PercentageOfLosing",value=percentage_of_losing,allow_duplicates=True)
```

```python
# Let's add a column WinningPerLosing the games
winning_per_losing=df["WonGames"]/df["LostGames"]
df.insert(loc=len(df.columns),column="WinningPerLosing",value=winning_per_losing,allow_duplicates=True)
```

```python
# Let's add a column BasketScoredPerGiven the games
basket_scored_per_given=df["BasketScored"]/df["BasketGiven"]
df.insert(loc=len(df.columns),column="BasketScoredPerGiven",value=basket_scored_per_given,allow_duplicates=True)
```

```python
# Lets import the necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
# Let's take some of the int64 columns to plot univariate graphs to check distribution of data
attributes=['Score','PlayedGames','WonGames','DrawnGames','LostGames','BasketScored','BasketGiven','TournamentChampion','Runner-up','PercentageOfWinning','PercentageOfLosing','WinningPerLosing','BasketScoredPerGiven']
plt.figure(figsize=(30,30))
attribute_index=1
for column_name in attributes:
    plt.subplot(5,3,attribute_index)
    sns.histplot(x=df[column_name],kde=True)
    attribute_index=attribute_index+1
```

```python
# Let's take some of the int64 columns against column Team to plot bivariate graphs to check the relationship between Team and other columns, individually
plt.figure(figsize=(20,50))
attribute_index=1
for column_name in attributes:
    plt.subplot(20,1,attribute_index)
    sns.barplot(x=df["Team"],y=df[column_name])
    plt.xticks(rotation=90)
    attribute_index=attribute_index+1
```

```python
import warnings   
# Settings the warnings to be ignored 
warnings.filterwarnings("ignore") 
# Let's check the correlation between
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True);
```
